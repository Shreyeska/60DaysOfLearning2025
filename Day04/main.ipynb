```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Gita'],
    'Age': [25, 30, 28, 35, 40, 29, 23],
    'Score': [85, 90, 88, 92, 95, 78, 81],
    'Gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F']
}
df = pd.DataFrame(data)
```

# Matplotlib Basics


```python
# Line Plot
plt.figure(figsize=(6, 4))
plt.plot(df['Name'], df['Score'], marker='o')
plt.title('Scores by Name')
plt.xlabel('Name')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Bar Chart
plt.figure(figsize=(6, 4))
plt.bar(df['Name'], df['Age'], color='orange')
plt.title('Age by Name')
plt.xlabel('Name')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Histogram
plt.figure(figsize=(6, 4))
plt.hist(df['Score'], bins=5, color='green', edgecolor='black')
plt.title('Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Countplot
plt.figure(figsize=(5, 4))
sns.countplot(x='Gender', data=df)
plt.title('Gender Count')
plt.tight_layout()
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(x='Gender', y='Score', data=df)
plt.title('Score Distribution by Gender')
plt.tight_layout()
plt.show()
```

```python
# Heatmap (Correlation)
plt.figure(figsize=(5, 4))
corr = df[['Age', 'Score']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
