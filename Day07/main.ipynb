```python
import matplotlib.pyplot as plt
import numpy as np
```

# Line Plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(6, 4))
plt.plot(x, y, label='sin(x)', color='blue')
plt.title("Line Plot: Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Bar Chart

```python
labels = ['Python', 'C++', 'Java', 'JavaScript']
popularity = [80, 65, 70, 75]

plt.figure(figsize=(6, 4))
plt.bar(labels, popularity, color='green')
plt.title("Programming Language Popularity")
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.tight_layout()
plt.show()
```

# Scatter Plot


```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)

plt.figure(figsize=(6, 4))
plt.scatter(x, y, c=colors, cmap='viridis', s=100, alpha=0.7, edgecolors='k')
plt.title("Random Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar(label='Color scale')
plt.tight_layout()
plt.show()
```
